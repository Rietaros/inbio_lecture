```python
import pandas as pd
```

```python
#read data actives
data_act = pd.read_csv("fitur_ff_pubchem_actives.csv", sep=",", header=0)
data_act.head()
```

```python
#add label actives
data_act['label'] = 1
data_act.head()
```

```python
#read data decoys
data_dec = pd.read_csv("fitur_ff_pubchem_decoys.csv", sep=",", header=0)
data_dec.head()
```

```python
#add label decoys
data_dec['label'] = 0
data_dec.head()
```

```python
#concat/merge data
frames = [data_act, data_dec]
data_all = pd.concat(frames)
data_all.head()
```

```python
#get feature and label only
data_model = data_all.iloc[:, 3:885]
data_model.head()
```

```python
#Train data with Logistic Regression, SVM, Random Forest
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#train test split 75%training 25% testing
X_train, X_test, y_train, y_test = train_test_split(data_model.loc[:, ~data_model.columns.isin(['label'])], data_model['label'],
                                                    test_size=0.3,random_state=8)
```

```python
#create model
lr_mdl = LogisticRegression()
svm_mdl = svm.SVC(C=1.0, kernel='rbf', probability = True)
rf_mdl = RandomForestClassifier(n_estimators=11)
```

```python
model_lr = lr_mdl.fit(X_train,y_train)
model_svm = svm_mdl.fit(X_train,y_train)
model_rf = rf_mdl.fit(X_train,y_train)
```

```python
#predict test data
y_pred_lr = model_lr.predict(X_test)
y_pred_svm = model_svm.predict(X_test)
y_pred_rf = model_rf.predict(X_test)
```

```python
#compute accuracy
print("Logistic Regression Accuracy:",metrics.accuracy_score(y_test, y_pred_lr))
print("SVM Accuracy:",metrics.accuracy_score(y_test, y_pred_svm))
print("Random Forest Accuracy:",metrics.accuracy_score(y_test, y_pred_rf))
```

```python
#Other Metrics
#Create and orint Confussion Matix
print("Logistic Regression: \n",metrics.confusion_matrix(y_test, y_pred_lr))
print("SVM: \n",metrics.confusion_matrix(y_test, y_pred_svm))
print("Random Forest: \n",metrics.confusion_matrix(y_test, y_pred_rf))
```

```python
#TP,FP,TN,FN example
confusion = metrics.confusion_matrix(y_test, y_pred_lr)
TP = confusion[1, 1]
TN = confusion[0, 0]
FP = confusion[0, 1]
FN = confusion[1, 0]

#Sensitivity / Recall = When the actual value is positive, how often is the prediction correct?
print(TP / float(TP + FN))

print("Logistic Regression: ",metrics.recall_score(y_test, y_pred_lr))
print("SVM: ",metrics.recall_score(y_test, y_pred_svm))
print("Random Forest: ",metrics.recall_score(y_test, y_pred_rf))
```

```python
#Specificity = When the actual value is negative, how often is the prediction correct?
print(TN / float(TN + FP))
```

```python
#Precision = When a positive value is predicted, how often is the prediction correct?
print(TP / float(TP + FP))

print("Logistic Regression: ",metrics.precision_score(y_test, y_pred_lr))
print("SVM: ",metrics.precision_score(y_test, y_pred_svm))
print("Random Forest: ",metrics.precision_score(y_test, y_pred_rf))
```

```python
#calculate ROC and draw
#Use predict_proba to get probability of positive

y_pred_prob_lr = model_lr.predict_proba(X_test)
y_pred_prob_svm = model_svm.predict_proba(X_test)
y_pred_prob_rf = model_rf.predict_proba(X_test)

y_prob_lr = y_pred_prob_lr[:, 1]
y_prob_svm = y_pred_prob_svm[:, 1]
y_prob_rf = y_pred_prob_rf[:, 1]
```

```python
# allow plots to appear in the notebook
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#calculate ROC and draw
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_prob_lr)
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC curve for Logistic Regression')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
print("AUC Logistic Regression: ",metrics.auc(fpr, tpr))
```

```python
#calculate ROC and draw
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_prob_svm)
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC curve for SVM')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
print("AUC SVM: ",metrics.auc(fpr, tpr))
```

```python
#calculate ROC and draw
fpr, tpr, thresholds = metrics.roc_curve(y_test, y_prob_rf)
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC curve for Random Forest')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
print("AUC Random Forest: ",metrics.auc(fpr, tpr))
```
