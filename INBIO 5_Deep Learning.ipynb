```python
import h2o
```

```python
h2o.init(max_mem_size = 2)            #uses all cores by default
h2o.remove_all()    
```

```python
from h2o.estimators.deeplearning import H2OAutoEncoderEstimator, H2ODeepLearningEstimator
help(H2ODeepLearningEstimator)
```

```python
#membaca file csv
df1 = h2o.import_file("fitur_ff_pubchem_actives.csv", header=1)
df2 = h2o.import_file("fitur_ff_pubchem_decoys.csv", header=1)

df2.head()
```

```python
#put label to data
df1['label'] = 1
df2['label'] = 0
```

```python
df1.head()
```

```python
#ganti tipe data
df1['code'] = df1['code'].ascharacter()
df1['label'] = df1['label'].asfactor()
df2['chembl_idx'] = df2['chembl_idx'].ascharacter()
df2['label'] = df2['label'].asfactor()
```

```python
#gabung dua data jadi satu
all_data = df1.rbind(df2)
```

```python
model_data = all_data[:,3:885]
model_data.head()
```

```python
## METHOD 2
# # Construct validation and training datasets by sampling (20/80)
iloc = model_data[0].runif()
train = model_data[iloc < 0.7]
valid = model_data[iloc >= 0.7]
```

```python
## get the shape of each of the datasets
print ('train shape', train.shape)
print ('validation shape', valid.shape)
```

```python
#mendapatkan nama kolom
predictors = all_data.col_names[3:-1]     #last column is response, our desired response variable 
response = all_data.col_names[-1]
```

```python
response
```

```python
from h2o.estimators.random_forest import H2ORandomForestEstimator
#https://h2o-release.s3.amazonaws.com/h2o/rel-turan/4/docs-website/h2o-py/docs/modeling.html#h2orandomforestestimator

from h2o.grid.grid_search import H2OGridSearch
## same as sklearn
## from sklearn.ensemble import GradientBoostingClassifier
```

```python
rf = H2ORandomForestEstimator(seed=1, 
    nfolds=5, #cross validation
    model_id="rf",
    ntrees=200,
    max_depth=30,
    stopping_rounds=2,
    stopping_tolerance=0.01,
    score_each_iteration=True)
```

```python
rf.train(x=predictors, y=response, training_frame=train, validation_frame = valid)
```

```python
#important variable
rf.varimp()
```

```python
print(rf.auc())
```

```python
print(rf.accuracy)
```

```python
dl_2 = H2ODeepLearningEstimator(hidden=[400,100], epochs=500, 
                               
  score_duty_cycle=0.025,          # don't score more than 2.5% of the wall time
  adaptive_rate=False,             # manually tuned learning rate
  rate=0.01, 
  rate_annealing=0.000002,            
  momentum_start=0.2,              # manually tuned momentum
  momentum_stable=0.4, 
                                variable_importances=True)
```

```python
dl_2.train(x=predictors, y=response, training_frame=train, validation_frame = valid)
```

```python
print(dl_2.accuracy)
```

```python
cov_tuned_df = dl_2.scoring_history()
cov_tuned_df['training_classification_error']
```

```python
pred = dl_2.predict(valid[:-1]).as_data_frame(use_pandas=True)
test_actual = valid.as_data_frame(use_pandas=True)['label']

(test_actual == pred['predict']).mean()
```
